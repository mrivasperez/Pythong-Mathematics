```python
x = 5
y = 5.1

ans1 = x**(3/4) * 4**y
ans2 = ((3**3) / x**y)
ans3 = 10**(x-4)
print(ans1)
print(ans2)
print(ans3)
```

```python
display(Math('5^{3/4} + 4^{5.1} = %g' %ans1) )
display(Math('\\frac{3^3}{5^{5.1}} = %g' %ans2) )
display(Math('10^{5-4} = %g' %ans3) )



```
