Printing out equations
Using LaTex to make them prettier

```python
from IPython.display import display, Math

display( '4 + 3 = 7' )
display( '4 + 3 = ' + str(4+3) )
```

```python
display(Math("4+3="+str(4+3)))
```

```python
x = 4
y = 5

display(Math(str(x)+ '+' + str(y) + '=' + str(x + y)))
```

```python
display(Math("%g + %g = %g" %(x, y, x + y)))
```

```python
display(Math(''))
# using LaTex to display fractions
display(Math("\\frac{4}{5} = .8"))
```

```python
x = 3.4
y = 17
ans = x * y
display(Math('%g \\times %g = %g' %(x,y, ans)))
```

```python
x = 7
y = -2
z =5

ans1 = (3 * x) * (4 + y)
display(Math('3 \\times %g(4 + %g)=%g' %(x,y, ans1)))

ans2 = -y - ((x+3)/z)
display(Math('--2 - \\frac{7+3}{5} = '+ str(ans2)))
```
